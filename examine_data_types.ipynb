```python
import json
import random
import pydicom as dicom
import matplotlib.pylab as plt
import matplotlib.image as mpimg
```

```python
def view(folder, img_id):
    
    path = f'dcm_data/{folder}/{img_id}'
    ds = dicom.dcmread(path)

    if ds.pixel_array.ndim == 3:
        plt.imshow(ds.pixel_array[:,:,:])
    elif ds.pixel_array.ndim == 4:
        print(ds.pixel_array.shape)
        i = random.randint(0, ds.pixel_array.shape[0]-1)
        plt.imshow(ds.pixel_array[i,:,:,:])
```

```python
with open("dcm_data_types_v2.json", "r") as f:
    data_type_dict = json.load(f)
```

```python
def check_type(type_i, ith_file):

    files = data_type_dict[str(type_i)]
    print(files[ith_file])
    view(files[ith_file].split("/")[0], files[ith_file].split("/")[1])
```
